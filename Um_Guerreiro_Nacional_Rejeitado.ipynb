```
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import pyplot as pls
import numpy as np
%matplotlib inline
```

```
temporadas_f1 = pd.read_csv('https://raw.githubusercontent.com/zackcmariano/Um-Guerreiro-Nacional-Rejeitado/main/temporadas_rubensbarrichello.csv')
```

```
temporadas_f1.head()
```

```
data = temporadas_f1.ANO
year = ['2000', '2001', '2002', '2003', '2004', '2005','2006','2007','2008','2009','2010','2011']
x = np.arange(len(data))

fig,ax = plt.subplots(figsize=(12,6))

Podiums = ax.bar(x, temporadas_f1.PODIUMS, width = 0.85, color = '#23807e')
#GP = ax.bar(x, temporadas_f1.GP, width = 0.40, color = '#6ae1e2')

plt.xticks(fontsize=14)
ax.set_title('SUBIU AO PODIUM', fontsize=20, color='#1f6376')
ax.set_ylabel('GP / Temporada', fontsize=15, color='#25424a')

ax.set_ylim([0,18])
ax.set_xticks(x)
ax.set_xticklabels(year)

plt.legend(labels=['Podiums'], fontsize=18);

def autolabel(system):
  for i in system:
    percentage = i.get_height()
    ax.annotate('{}'.format(percentage),
                xy = (i.get_x()+i.get_width()/2,percentage),
                xytext = (0,5),
                textcoords = 'offset points',
                ha = 'center', fontsize=16)    

autolabel(Podiums)
#autolabel(GP)

#plt.savefig('BrowsersBr.png',dpi = 600)   / TO SAVE PNG

plt.show()
```
